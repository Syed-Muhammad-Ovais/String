```python
##Python String Concatenation and Formatting Assignments
#Assignment 1: String Concatenation and Formatting
name = "Syed Muhammad Ovais"
age = 38
country = "Qatar"
detail = "My name is " + name + " I am " + str(age) + " years old, I live in " + country + "."
print(detail)
```

```python
#Assignment 2: String Formatting using `.format()
product = "laptops"
price = 500
quantity = 10
total_cost = price*quantity
con = "I bought {} {} at ${} each, for a total of ${}"
summary = con.format(quantity, product, price, total_cost)
print(summary)
```

```python
## Assignment 3: String Interpolation using `f''`
city = "Doha"
temperature = 45
sent = "The temperature in {} is {}°C"
temp=sent.format(city,temperature)
print (temp)
```

```python
## Assignment 4: String Formatting using
first_name = "Syed"
last_name = "Ovais"
birth_year = 1985
current_year = 2023
age = current_year - birth_year
data = "My name is %s %s  I am %d years old. " % (first_name,last_name,age)
print (data)
```

```python
##Assignment 5: Combining Formatting Methods

product = "Playstation 5"
discount = 10
original_price = 400
discounted_price = original_price *(1 - discount / 100)
discounted_percentage = (discount / original_price)*100
detail = ("Sale on {} at {}% discount, original price was {}$, after discount {}$ price.")
sum = detail.format(product,discounted_percentage,original_price,discounted_price)
sum
```
